```python
#importar la libreria numpy, que es la que vamos a utilizar para operar con las operaciones matematicas
import numpy as np
import matplotlib.pyplot as plt #importamos la libreria matplotlic para la definicon de las funciones
import math 
```

```python
#Definimos las funciones
def f(x):
    return pow (x, 2)-x+1 #(funcion 1)

g= lambda x: (2)/(x-1) #(funcion 2)

```

```python
#Damos valores al rango de la grafica (parametros)
x=np.linspace(-3,4,58) #parametros
plt.plot(x,f(x),label= 'Funcion x')
plt.plot(x,g(x),label= 'Funcion g')
plt.xlabel("Eje x")
plt.ylabel("Eje y")
plt.grid() #cuadricula
```
